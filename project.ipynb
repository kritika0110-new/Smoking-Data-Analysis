```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
file_path = "/content/datafile.csv"  # Ensure your file is uploaded to this path
df = pd.read_csv(file_path)

# Select only the required numerical columns
features = [
    "Current tobacco smokers (%)",
    "Current cigarette users (%)",
    "Current bidi users (%)",
    "Students who noticed anti-tobacco messages in mass media in past 30 days    (%)",
    "Students who were taught in class about harmful effects of tobacco use during past 12 months (%)",
    "Students who favoured ban on smoking inside enclosed public places (%)",
    "Schools followed ‘tobacco-free school’ guidelines  (%)"
]

# Group by State/UT and take the average values
df_grouped = df.groupby("State/UT")[features].mean().reset_index()

# Normalize the numerical data (scale between 0 and 1)
scaler = MinMaxScaler()
df_grouped[features] = scaler.fit_transform(df_grouped[features])

# Function to draw a Chernoff-like face
def draw_face(ax, data, state_name):
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1.2, 1.2)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_frame_on(False)

    # Face shape
    face_width = 0.6 + data[0] * 0.4  # Based on Current tobacco smokers (%)
    face_height = 0.8 + data[5] * 0.4  # Based on Students favoring tobacco ban
    face = plt.Circle((0, 0), 0.8, color="peachpuff", ec="black")
    ax.add_patch(face)

    # Eyes
    eye_size = 0.05 + data[1] * 0.05  # Based on Current cigarette users (%)
    ax.add_patch(plt.Circle((-0.3, 0.4), eye_size, color="black"))
    ax.add_patch(plt.Circle((0.3, 0.4), eye_size, color="black"))

    # Nose
    nose_height = 0.05 + data[2] * 0.1  # Based on Current bidi users (%)
    ax.plot([0, -0.05], [0.1, -nose_height], "k", lw=2)
    ax.plot([0, 0.05], [0.1, -nose_height], "k", lw=2)

    # Mouth
    mouth_curve = -0.2 + data[4] * 0.4  # Based on Students learning about tobacco harm
    ax.plot([-0.3, 0, 0.3], [-0.4, mouth_curve, -0.4], "r", lw=2)

    # Eyebrows (based on tobacco ads exposure)
    eyebrow_angle = data[3] * 0.3  # Based on Students exposed to tobacco ads
    ax.plot([-0.35, -0.15], [0.55 + eyebrow_angle, 0.55], "k", lw=2)
    ax.plot([0.35, 0.15], [0.55 + eyebrow_angle, 0.55], "k", lw=2)

    # Add State/UT label
    ax.set_title(state_name, fontsize=10)

# Plot faces for unique states (adjustable count)
num_faces = min(9, len(df_grouped))  # Show max 9 states
fig, axes = plt.subplots(3, 3, figsize=(10, 10))
axes = axes.flatten()

for i in range(num_faces):
    draw_face(axes[i], df_grouped.iloc[i, 1:].values, df_grouped["State/UT"].iloc[i])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


# Select the columns you want to use as features
features = ['Current tobacco smokers (%)' ,'Current cigarette users (%)' ,'Current bidi users (%)']

# Select the column you want to predict
target = 'Ever tobacco smokers who quit in last 12 months (%)'

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2)

# Create and fit the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f'Mean squared error: {mse:.2f}')

# Show the coefficients for each feature
coefficients = pd.DataFrame({'Feature': features, 'Coefficient': model.coef_})
print(coefficients)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.feature_selection import SelectKBest, f_regression




features = ['Current tobacco smokers (%)' ,'Current cigarette users (%)' ,'Current bidi users (%)']
target = 'Ever tobacco smokers who quit in last 12 months (%)'


X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2)

# Select the k best features using f_regression as the scoring function
k = 10
selector = SelectKBest(f_regression, k=k)
X_train_new = selector.fit_transform(X_train, y_train)
X_test_new = selector.transform(X_test)

model = LinearRegression()
model.fit(X_train_new, y_train)


y_pred = model.predict(X_test_new)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f'Mean squared error: {mse:.2f}')

# Show the selected features
selected_features = [features[i] for i in selector.get_support(indices=True)]
print(f'Selected features: {selected_features}')
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error


# Select the columns you want to use as features
features = ['Current tobacco smokers (%)' ,'Current cigarette users (%)' ,'Current bidi users (%)']
# Select the column you want to predict
target = 'Ever tobacco smokers who quit in last 12 months (%)'

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2)

# Define the models to evaluate
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree Regression': DecisionTreeRegressor(),
    'Random Forest Regression': RandomForestRegressor(),
    'Gradient Boosting Regression': GradientBoostingRegressor()
}

# Fit and evaluate each model
for name, model in models.items():
    # Fit the model
    model.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = model.predict(X_test)

    # Calculate the mean absolute error (MAE)
    mae = mean_absolute_error(y_test, y_pred)

    # Calculate the root mean squared error (RMSE)
    mse = (mean_squared_error(y_test, y_pred))

    # Print the results
    print(f'{name}:')
    print(f'  MAE: {mae:.2f}')
    print(f'  MSE: {mse:.2f}')
```
